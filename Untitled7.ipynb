```python
import os
import csv
import random
f ="/content/data/movie-dialog-corpus/movie_lines.tsv"
texts = []
with open(f) as i:
    csv_reader = csv.reader(i, delimiter="\t")
    for line in csv_reader:
        texts.append(line[-1])

random.seed(1)
random.shuffle(texts)
texts = texts[:10000]
texts[:10]
```

```python
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
embed=hub.load("https://tfhub.dev/google/universal-sentence-encoder/2")
```

```python
def embed_texts(texts):
    # Ensure input is a TensorFlow tensor of strings
    input_tensor = tf.constant(texts, dtype=tf.string)

    # The 'embed' object, for this specific module version, is a tf.Module
    # that is not directly callable. Instead, its embedding functionality
    # is exposed through its 'call' method.
    embeddings_tensor = embed.call(input_tensor)

    # Convert the tf.Tensor to a numpy array, then to a list
    embeddings = embeddings_tensor.numpy()

    return embeddings.tolist()

embeddings = embed_texts(texts)
```

```python
import tensorflow_hub as hub
import tensorflow as tf

# Load a text embedding model (example: Universal Sentence Encoder)
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

def embed_texts(texts):
    # Ensure input is a TensorFlow tensor of strings
    input_tensor = tf.constant(texts)

    # Call the module directly
    embeddings_tensor = embed(input_tensor)

    # Convert to numpy, then to list
    embeddings = embeddings_tensor.numpy()
    return embeddings.tolist()

# Example usage
texts = ["Hello world", "Embeddings are powerful"]
embeddings = embed_texts(texts)
print(len(embeddings), len(embeddings[0]))  # number of texts, embedding dimension
```
