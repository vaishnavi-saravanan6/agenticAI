```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
```

```python
df=pd.read_csv("/content/salary_dataset (1).csv")
```

```python
X=df.drop(['Salary','Unnamed: 0'],axis=1)
Y=df["Salary"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model=LinearRegression()
```

```python
model=model.fit(X_train,Y_train)
```

```python
predictedSalary=model.predict(X_test)
```

```python
predictedSalary
```

```python
errors=mean_squared_error(Y_test,predictedSalary)
print("Mean Squared Error:",errors)

r2=r2_score(Y_test,predictedSalary)
print("R-squared:",r2)

mae=mean_absolute_error(Y_test,predictedSalary)
print("Mean Absolute Error:",mae)

mape=mean_absolute_percentage_error(Y_test,predictedSalary)
print("Mean Absolute Percentage Error:",mape)
```
